```python
from math import radians, cos, sin, asin, sqrt
import keras
from keras.models import Sequential 
from keras.layers import Dense
from keras.layers import Dropout 
from keras.optimizers import Adam
import tensorflow as tf
from tensorflow import keras
```

```python
df3=df3.loc[1:250000]
len(df3)
```

```python
Features=['VendorID','passenger_count',
              'trip_distance','RatecodeID','PULocationID','DOLocationID','payment_type','fare_amount',
              'extra','mta_tax','tip_amount','tolls_amount','improvement_surcharge','total_amount','congestion_surcharge','Days',
              'tavg','tmin','tmax','wspd','pres']
Target=['trip_duration']     
```

```python
x = df3[Features].values
y = df3[Target].values
from sklearn.preprocessing import StandardScaler
Fscalar=StandardScaler()
Tscalar=StandardScaler()


FeaturesFit=Fscalar.fit(x)
TargetFit=Tscalar.fit(y)

x=FeaturesFit.transform(x)
y=TargetFit.transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1337)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

# MLP

```python
model = Sequential()
```

```python
model.add(Dense(units=64, input_dim=21, activation='relu'))
```

```python
model.add(Dense(units=32, activation='relu'))
```

```python
model.add(Dropout(0.5))
```

```python
model.add(Dense(units=16, activation='relu'))
```

```python
model.add(Dropout(0.5))
```

```python
model.add(Dense(units=1,activation='linear'))
model.summary()
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=20)
```

### Adam MLP

```python
model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics=['mae'])
history = model.fit(x_train, y_train, epochs=100, validation_data=(x_test, y_test), batch_size = 32, verbose=1,shuffle=True, callbacks=[callback])
```

```python
predict = model.predict(x_test)
```

```python
Predict=model.predict(x_test)
predict=TargetFit.inverse_transform(predict)
y_test_original = TargetFit.inverse_transform(y_test)
x_test_original = FeaturesFit.inverse_transform(x_test)
PredictedData=pd.DataFrame(data=x_test_original, columns=Features)
PredictedData['trip_duration']=y_test_original
PredictedData['Predicted']=predict
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('val_loss')
plt.legend(['Train','Test'], loc='upper right')
plt.show()
```
